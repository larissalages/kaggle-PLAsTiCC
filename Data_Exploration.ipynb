```python
#import everything you need

import pandas as pd
from datetime import datetime
```

```python
#Define functions 

def convert_MDJ2unix(df):
    return (df -40587)*86400

def timestamp2readbleTime(time):
    return datetime.utcfromtimestamp(time).strftime('%Y-%m-%d %H:%M:%S')
```

```python
#read csv
df_1 = pd.read_csv('training_set.csv')
df_2 = pd.read_csv('training_set_metadata.csv')
```

```python
#Convert to Unix timestamp
df_1.mjd = convert_MDJ2unix(df_1.mjd)
```

```python
#Convert to readable time
# Put the results in a list and after the loop finishes assigning the list to the column of the dataframe accelerates the assignment in dataframes

list_time = []

for i in range(len(df_1.mjd)):
    list_time.append(timestamp2readbleTime(df_1.mjd[i]))
    
df_1.mjd = list_time
```
